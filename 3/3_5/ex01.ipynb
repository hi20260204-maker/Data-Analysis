# 주식 데이터 다운로드

```python
import yfinance as yf

import pandas as pd

#Apple 주식 데이터
apple = yf.Ticker("AAPL")

#다운로드
df = apple.history(period = '1y')
df.shape #(전체 데이터 수, 전체 변수 수)
```

```python
# 데이터 구경하자
df.info()
```

```python
df.head()
```

```python
df_in = pd.DataFrame(df.index, dtype="str")
```

```python
'2025-03-05 00:00:00-05:00'.split(" ")[0]

# df의 인덱스를 Date로 변경
df = df.reset_index()

# Date값을 str로 다시 넣고 필요한 부분만 남겨두기
df["Date"] = [
    str(df["Date"][i]).split(" ")[0] for i in range(len(df))
]

#df에서 Dividends, Stock Splits 제외
df = df[["Date","Open","High","Low","Close","Volume"]]
df.head()
```

```python
# df에서 새로운 컬럼 추가 및 생성
# 년, 월, 일, 변화량 => 생성
df["연"] = df["Date"].apply(lambda x: x.split("-")[0])
df["월"] = df["Date"].apply(lambda x: x.split("-")[1])
df["일"] = df["Date"].apply(lambda x: x.split("-")[2])
df["변화량"] = df["Close"] - df["Open"]
df
```

```python
# 연, 월, 일, 변화량 컬럼으로만 데이터프레임 생성
df1 = df[["연","월","일","변화량"]]
df1.head()
```

```python
df.rename(columns={"Date":"날짜"})
```

```python
# 볼린저 밴드의 기간
window = 20

# rolling => row 데이터에 대한 연산
df["sma_20"] = df["Close"].rolling(window=window).mean()
df["std_20"] = df["Close"].rolling(window=window).std()

df.tail()
```

```python
df['bollinger_up']=df.apply(lambda x: x['sma_20'] + x['std_20'] * 2, axis=1)
df['bollinger_down']=df.apply(lambda x: x['sma_20'] - x['std_20'] * 2, axis=1)
df.tail()
```

```python
# 주식 그래프 그리기 
import plotly.graph_objects as go
from plotly.subplots import make_subplots

fig = make_subplots(specs=[[{"secondary_y": True}]])
fig.add_trace(go.Scatter(x=df['Date'], y=df['Close'], mode='lines', name='종가', line_color="green"), secondary_y=False)

fig.add_trace(go.Scatter(x=df['Date'], y=df['bollinger_up'], mode='lines', name='상한선', line_color="red"), secondary_y=False)
fig.add_trace(go.Scatter(x=df['Date'], y=df['bollinger_down'], mode='lines', name='하한선', line_color="blue"), secondary_y=False)

fig.update_layout(xaxis_rangeslider_visible=False)
fig.show()
```

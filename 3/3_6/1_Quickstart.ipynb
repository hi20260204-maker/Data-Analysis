```python
import easydict # 주피터에서 변수들을 저장하는 객체
```

```python
args = easydict.EasyDict() # 객체 생성

#변수 저장
args.target_col = 'target' # 타겟 컬럼명
```

```python
args.target_col
```

# 텍스트 수집

```python
from sklearn.datasets import load_breast_cancer

# 유방암 데이터셋
data = load_breast_cancer()

dir(data)
```

```python
data.feature_names
```

```python
data.target_names # 예측해야 하는 변수
```

# 데이터 분석(EDA)

```python
import numpy as np
import pandas as pd

df = pd.DataFrame(data.data, columns=data.feature_names) # feature 데이터만...
df[args.target_col] = data[args.target_col] #target 추가...

df.shape
```

```python
df.info()
```

```python
df.describe()
```

# 학습용 데이터와 평가용 데이터 분리

```python
from sklearn.model_selection import train_test_split

df_train, df_test = train_test_split(df)

df_train.shape, df_test.shape
```

# 모델 훈련

```python
from sklearn.linear_model import LogisticRegression # 이진 분류 모델

model = LogisticRegression() # 모델 선언
```

```python
# feature & target 학습 진행 -> 지도 학습

y_train = df_train[args.target_col] #target
x_train = df_train.drop(args.target_col, axis=1) #feature

model.fit(x_train, y_train) # 모델 학습
```

```python
from sklearn.metrics import accuracy_score # 평가 지표 -> 지금은 정확도

y_test = df_test[args.target_col] #target(정답)
x_test = df_test.drop(args.target_col, axis=1) #feature

pred = model.predict(x_test) #테스트용 feature 이용해서 예측값(preq) 생성
```

```python
accuracy_score(y_test, pred) #정답과 예측값 비교
```

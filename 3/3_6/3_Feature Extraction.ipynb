# 진행바
> 프로그램(&모델 학습시간..)이 얼마나 진행되고 있는지 표시하는 라이브러리

```python
from tqdm.auto import tqdm
```

```python
i = 0
for j in tqdm(
    np.random.rand(100000000)
    , desc ="설명글", leave=False
    ):
  i = i**2
```

```python
import time

# outer => 진행바가 유지(leave =True)
for outer in tqdm([10,20,30,40,50], desc="outer", leave=True):
  # inner => 진행바가 유지(leave = False)
  for inner in tqdm(range(100), desc="inner", leave=False):
    time.sleep(0.01)
```

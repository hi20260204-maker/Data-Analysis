> 전체 데이터에서 train과 test를 나누는 기준은 무엇인가<br>
> if 전체 데이터가 충분히 많으면 train : test = 7: 3<br>
> if 전체 데이터가 충분하지 않으면 train : test = 9: 1<br>
> if 평가 데이터를 충분히 정말 충분히 얻을 수 없다면...<br>
> 기억하자!! test의 비율을 낮출 수록 당연히 모델의 퀄리티는 다운된다.

```python
from sklearn.model_selection import  train_test_split

train, test = train_test_split(df, test_size=0.18)

train.shape, test.shape
```

```python
train.reset_index(drop = True, inplace = True)
test.reset_index(drop = True, inplace = True)
```

# Feature Extraction

```python
train.head()
```

```python
df_number = train.select_dtypes(np.number)
df_number.columns
```

```python
df_number.info()
```

```python
df_number['survived'] = df_number['survived'].astype('int32')
df_number['pclass'] = df_number['pclass'].astype('int32')
df_number['age'] = df_number['age'].astype('float32')
df_number['is_cabin'] = df_number['is_cabin'].astype('float32')

df_number.info()
```

```python

train['survived'] = train['survived'].astype('int32')
train['pclass'] = train['pclass'].astype('int32')
train['age'] = train['age'].astype('float32')
train['is_cabin'] = train['is_cabin'].astype('float32')

test['survived'] = test['survived'].astype('int32')
test['pclass'] = test['pclass'].astype('int32')
test['age'] = test['age'].astype('float32')
test['is_cabin'] = test['is_cabin'].astype('float32')

train.info()
```

```python
# 절대 test를 보지 말아야함
train['name'].head()
```

```python
train['name'] = train['name'].map(lambda x: x.strip())
test['name'] = test['name'].map(lambda x: x.strip())

train['name'].head()
```

```python
"Mr." in "Sivic, Mr. Husein"
```

```python
dict_designation = {
    'Mr.': '남성',
    'Master.': '남성',
    'Sir.': '남성',
    'Miss.': '미혼 여성',
    'Mrs.': '기혼 여성',
    'Ms.': '미혼/기혼 여성',
    'Lady.': '숙녀',
    'Mlle.': '아가씨',
    # 직업
    'Dr.': '의사',
    'Rev.': '목사',
    'Major.': '계급',
    'Don.': '교수',
    'Col.': '군인',
    'Capt.': '군인',
    # 귀족
    'Mme.': '영부인',
    'Countess.': '백작부인',
    'Jonkheer.': '귀족'
}
```

```python
def add_designation(name):
  #초기값
  result = "unknown"                  # 정의된 호칭이 없는 때, 모름(unknown)
  for key in dict_designation.keys(): # 정의된 호칭 정보들
    if key in name:                   # 만약, 사용자 이름에 호칭이 있다면,
      result = dict_designation[key]  # 해당 호칭 한국어명 리턴
      break
  return result

train['designation'] = train['name'].map(lambda x: add_designation(x))
test['designation'] = test['name'].map(lambda x: add_designation(x))

train[['name','designation']].head(100)
```
